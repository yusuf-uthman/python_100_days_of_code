## 1 coingecko
#### Leading Crytocurrency price listings

```python
url1 = 'https://api.coingecko.com/api/v3/exchange_rates'
response1  = requests.get(url1)
```

##### Approach 1

```python
if response1.status_code == 200:
    data_1 = response1.json()
    data_1 = data_1['rates']
    print(type(data_1))
    df =  pd.DataFrame(data_1)
    df = df.transpose() # df.T    #This can also transpose
else:
    print('Error with url')

# type(data_1)
# data
df.head(10)
```

##### Aproach 2

```python
if response1.status_code == 200:
    data_1 = response1.json()
    data_1 = data_1['rates']

    keys = []
    for val in data_1:
        keys.append(val)
# print(keys)

    data_coll = []
    for val in keys:
        data = data_1[val]
        data_coll.append(data)
        df_data_coll =  pd.DataFrame(data_coll)
else:
    print('Error with the api')
```

##  5 genderize.io
#### A simple API to predict the gender of a person given their name

```python
name = input("Tell is your name : \n")
```

```python
base_url = 'https://api.genderize.io?name='
full_url = f"{base_url}{name}"
response5 = requests.get(full_url)
if response5.status_code == 200:
    data = response5.json()
    name = data['name']
    gender = data['gender']
    probability = data['probability']
    count = data['count']
    print(f"""Hello {name},
    
We believe you are a {gender} cos we know {count} other people named {name} and they are all {gender}s.

Thank You!""")
```

```python
def gender_prediction(name):
    base_url = 'https://api.genderize.io?name='
    full_url = f"{base_url}{name}"
    response5 = requests.get(full_url)
    if response5.status_code == 200:
        data = response5.json()
        name = data['name']
        gender = data['gender']
        probability = data['probability']
        count = data['count']
        msg = print(f"""Hello {name},

        We believe you are a {gender} cos we know {count} other people named {name} and they are all {gender}s.

        Thank You!""")
    return msg
```
